```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Filter data
#Only get USA races, 50km or 50 Miles, 2020
```

```python
#combine 50mi and 50 km using isin
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020)].head(5)
```

```python
df[df['Event name'] == 'Avalon Benefit 50-Mile Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
#Get all the events at the USA
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'].head(5)
```

```python
#combine all filters
Filtered_df = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
Filtered_df.head(3)
```

```python
Filtered_df.shape
```

```python
#Remove (USA) from event name
Filtered_df['Event name'] = Filtered_df['Event name'].str.split('(').str.get(0)
```

```python
Filtered_df.head(3)
```

```python
#clean athlete age
Filtered_df['Athlete age'] = 2020 - Filtered_df['Athlete year of birth']
```

```python
#remove h from athlete's performance
Filtered_df['Athlete performance'] = Filtered_df['Athlete performance'].str.split(' ').str.get(0)
```

```python
Filtered_df.head(3)
```

```python
#drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
Filtered_df = Filtered_df.drop(['Athlete club','Athlete country','Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
Filtered_df.head()
```

```python
#cleanup null values
Filtered_df.isna().sum()
```

```python
Filtered_df.shape
```

```python
Filtered_df = Filtered_df.dropna()
```

```python
Filtered_df.shape
```

```python
#check for dupes
Filtered_df[Filtered_df.duplicated() == True]
```

```python
#reset Index
Filtered_df.reset_index(drop = True)
```

```python
#change data types
Filtered_df.dtypes
```

```python
Filtered_df['Athlete age']=Filtered_df['Athlete age'].astype(int)
```

```python
Filtered_df['Athlete average speed']=Filtered_df['Athlete average speed'].astype(float)
```

```python
Filtered_df.dtypes
```

```python
Filtered_df.head()
```

```python
#rename the columns
Filtered_df = Filtered_df.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_date',                 
    'Event name': 'race_name',                   
    'Event distance/length': 'race_distance',         
    'Event number of finishers': 'race_number_of_finishers',     
    'Athlete performance': 'athlete_performance',           
    'Athlete gender': 'athlete_gender',               
    'Athlete average speed': 'athlete_average_speed',        
    'Athlete ID': 'athlete_id',                    
    'Athlete age': 'athlete_age'                   
    
})
```

```python
Filtered_df.head()
```

```python
#drop the year column
Filtered_df = Filtered_df.drop(['year'], axis = 1)
```

```python
#charts and graphs
sns.histplot(Filtered_df['race_distance'])
```

```python
sns.histplot(Filtered_df, x ='race_distance', hue = 'athlete_gender')
```

```python
sns.displot(Filtered_df[Filtered_df['race_distance'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = Filtered_df, x = 'race_distance', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart')
```

```python
#Difference in speed for the 50km,50 mi male to female
Filtered_df.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#which age group is the best at the 50mi races(+100 athletes)
Filtered_df.query('race_distance == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean' , 'count']).sort_values('mean', ascending = False).query('count > 99')           
```

```python
#which age group is the worst at the 50mi races(+10 athletes)
Filtered_df.query('race_distance == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean' , 'count']).sort_values('mean', ascending = True).query('count > 9')               
```

```python
#Are runners slower in the summer or winter
#spring --> 3-5
#summer --> 6-8
#fall --> 9-11
#winter --> 12-2
#split between two decimals
```

```python
Filtered_df['race_month'] = Filtered_df['race_date'].str.split('.').str.get(1).astype(int)
```

```python
Filtered_df['race_season'] = Filtered_df['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
Filtered_df.groupby(['race_season'])['athlete_average_speed'].agg(['mean' , 'count']).sort_values('mean', ascending = False)
```
